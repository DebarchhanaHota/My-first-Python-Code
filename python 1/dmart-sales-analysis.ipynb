```python
train = pd.read_csv("/kaggle/input/dmart-sales/Dmart sales analysis/train (1).csv") # edit input path accordingly
train.head()
```

```python
test = pd.read_csv("/kaggle/input/dmart-sales/Dmart sales analysis/test (1).csv")
test.head()
```

```python
train.info()
```

```python
test.info()
```

```python
train1 = train.drop(['Item_Weight','Outlet_Size','Item_Identifier'],axis=1)
train2 = train.drop(['Item_Weight','Outlet_Size'],axis=1)
test1 = test.drop(['Item_Weight','Outlet_Size','Item_Identifier'],axis=1)
```

```python
train1['Item_Fat_Content'].unique().tolist()
```

```python
train1.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
train1.Item_Type= le.fit_transform(train1.Item_Type.values)
train1.Item_Fat_Content= le.fit_transform(train1.Item_Fat_Content.values)
train1.Outlet_Identifier= le.fit_transform(train1.Outlet_Identifier.values)
train1.Outlet_Location_Type= le.fit_transform(train1.Outlet_Location_Type.values)
train1.Outlet_Type= le.fit_transform(train1.Outlet_Type.values)
train1.head()
```

```python
test1.head()
```

```python
test1.Item_Type= le.fit_transform(test1.Item_Type.values)
test1.Item_Fat_Content= le.fit_transform(test1.Item_Fat_Content.values)
test1.Outlet_Identifier= le.fit_transform(test1.Outlet_Identifier.values)
test1.Outlet_Location_Type= le.fit_transform(test1.Outlet_Location_Type.values)
test1.Outlet_Type= le.fit_transform(test1.Outlet_Type.values)
test1.head()
```

```python
train1.info()
```

```python
test1.info()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

df_train, df_test = train_test_split(train1, test_size=0.2, random_state=100)

X_train = df_train.drop(columns=['Item_Outlet_Sales'])
y_train = df_train['Item_Outlet_Sales']
X_test = df_test.drop(columns=['Item_Outlet_Sales'])
y_test = df_test['Item_Outlet_Sales']

model1 = LinearRegression()

model1.fit(X_train, y_train)

test_predictions = model1.predict(X_test)
rmse = mean_squared_error(y_test, test_predictions, squared=False)
mae = mean_absolute_error(y_test, test_predictions)
r2 = r2_score(y_test, test_predictions)

print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"R-squared (R2): {r2:.2f}")

# model2 = LogisticRegression()

# model2.fit(X_train, y_train)

# test_predictions = model2.predict(X_test)
# rmse = mean_squared_error(y_test, test_predictions, squared=False)
# mae = mean_absolute_error(y_test, test_predictions)
# r2 = r2_score(y_test, test_predictions)

# print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
# print(f"Mean Absolute Error (MAE): {mae:.2f}")
# print(f"R-squared (R2): {r2:.2f}")
```

```python
test_pred = model1.predict(test1)
```

```python
prediction_df = pd.DataFrame({'Item_Identifier': test['Item_Identifier'], 'Item_Outlet_Sales': test_pred})

prediction_df.index = range(len(prediction_df))

prediction_df.to_csv('pred.csv', index=False)

print("Predictions saved to 'pred.csv'.")
```
